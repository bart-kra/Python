```python
import pandas as pd
```

```python
def computing(m, h, l, part):
  m_fac = [6.94, 43.457, 1.707, 1.625, 0.614, 14.165, 4.33, 1.371]
  rad_fac = [50, 44.5, 45, 42.7, 37, 45.5, 40.5, 44.1]

  df = pd.DataFrame()

  df['segment'] = part
  df['l_segment'] = l
  df['m_fac'] = m_fac 
  df['rad_fac'] = rad_fac 
  df['m_rel'] = df['m_fac'] * m * 0.01
  df['r'] = df['segment'] * df['rad_fac'] * 0.01

  return df
```

```python
# part = ['glowa', 'tulow', 'ramie', 'przedramie', 'dlon', 'udo', 'podudzie', 'stopa']

# print('Masa ciaÅ‚a: ', end='')
# m = int(input())
# print('Wzrost: ', end='')
# h = int(input())

# l = []

# for i in part:
#   print('l_{}: '.format(i), end='')
#   l.append(int(input()))

df = computing(m, h, l, part)
```

```python
df[['m_rel', 'r']] = df[['m_rel', 'r']].apply(lambda x: round(x, 2))
```

```python
itrt = 1
m_list = []
for i, j, k in zip(df['part'], df['m_rel'], df['r']):
  print('{}: \nm_{} = {}'.format(i, itrt, round(j, 2)))
  print('r_{} = {}\n'.format(itrt, round(k, 2)))
  itrt += 1
  m_list.append(round(j, 2))
print(m_list)
```

```python
r_list_x = [161,116.5,142,141.5,141,74.3,35,5.5]
r_list_y = [9,12.5,14,41.5,64.5,10,8,11]

df1 = pd.DataFrame()
df1['part'] = df['part']
df1['r_x1'] = r_list_x
df1['r_y1'] = r_list_y
df1['m_rel'] = df['m_rel']
df1['seg_count'] = [1, 1, 2, 2, 2, 2, 2, 2]
df1['m_rel*r_x1*seg_count'] = df1['r_x1'] * df1['m_rel'] * df1['seg_count']
df1['m_rel*r_y1*seg_count'] = df1['r_y1'] * df1['m_rel'] * df1['seg_count']

df1
```

```python
liczx = round(df1['m_rel*r_x1*seg_count'].sum(),2)
mianx = round(df1['m_rel'].sum(), 2)
print('x1 = {}'.format(liczx/mianx))
liczy = round(df1['m_rel*r_y1*seg_count'].sum(),2)
miany = round(df1['m_rel'].sum(), 2)
print('y1 = {}'.format(liczy/miany))
```
