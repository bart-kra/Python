```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

def ball_flight_no_drag(x0 = 0, z0 = 0, g = 0, v0 = 0, alfa = 0):
  list_x = []
  list_z = []
  alfa_rad = np.pi/180 * alfa
  z = z0
  for t in range(0,100):
    '''itterates through time steps'''
    if z >= 0:
      x = x0 + ((v0 * math.cos(alfa_rad)) * t)
      z = (z0 + ((v0 * math.sin(alfa_rad)) * t)) - ((g  * t ** 2) / 2)
      list_x.append(x)
      list_z.append(z)
  list_x.pop(-1)
  list_z.pop(-1)
  print('alfa = {}, V0 = {}'.format(alfa, v0))
  plotter(list_x, list_z, alfa, 'no drag')
  print('Length of a throw: {}'.format(round(list_x[-1]), 2))

def ball_flight_no_drag_no_plot(x0 = 0, z0 = 0, g = 0, v0 = 0, alfa = 0):
  list_x = []
  list_z = []
  alfa_rad = np.pi/180 * alfa
  z = z0
  for t in range(0,100):
    '''itterates through time steps'''
    if z >= 0:
      x = x0 + ((v0 * math.cos(alfa_rad)) * t)
      z = (z0 + ((v0 * math.sin(alfa_rad)) * t)) - ((g * t ** 2) / 2)
      list_x.append(x)
      list_z.append(z)
  list_x.pop(-1)
  list_z.pop(-1)
  return list_x[-1]

def plotter(data_x, data_z, alfa, drag):
  fig, ax = plt.subplots()
  ax.plot(data_x, data_z)
  ax.spines['left'].set_position('zero')
  ax.spines['bottom'].set_position('zero')
  plt.title(str(alfa)+'deg, '+str(drag))
  plt.xlabel('X')
  plt.ylabel('Z')

def best_deg():
  max_dist = 0
  opt_deg = 0
  dist = 1
  for deg in range(1, 90):
    dist = ball_flight_no_drag_no_plot(x0 = 0, z0 = 0, g = 10, v0 = 100, alfa = deg)
    if dist > max_dist:
      max_dist = dist
      opt_deg = deg
    print('{}. Most optimal deg is {} with distance {}'.format(deg, opt_deg, dist))
  
```
